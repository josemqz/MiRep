```python
import numpy as np
from scipy import interpolate
import pandas as pd
```

```python
XY = np.array([0, 0],
 [1, 0.70807341827357],
 [2, 0.82682181043181],
 [3, 0.019914856674817], 
 [4, 0.57275001690431], 
 [5, 0.91953576453823])

for i in XY.shape[0]:
    x_i = XY[i,0]
    y_i = XY[i,1]
    
    Si[i] = y_i + a_i(x − x_i) + b_i(x − x_i)**2 + c_i(x − x_i)**3
    
    Si = y_i + a_i*x - a_i*x_i + b_i(x**2 - 2*x*x_i + x_i**2) + c_i(x**3 - 3*x**2*x_i + 3*x*x_i**2 - x_i**3)
    Si_d(x) = a_i + 2*x*b_i - 2*x_i*b_i + 3*x**2*c_i - 6*x*x_i*c_i + 3*x_i**2*c_i
    
    
```

```python
errrange = np.linspace(2,10,500)

zeta = lambda s: np.sum([i**(-s) for i in range(1,101)])

N = 1
minErr = 1
NminErr = minErr
rminErr = errrange[0]
while N <= 70 and minErr > 10**(-12):
    
    Nrange = np.linspace(2,10,N)

    y_N = [zeta(Nrange[j]) for j in range(N)]

    IP = interpolate.BarycentricInterpolator(Nrange,y_N)
    
    
    #buscar maximo error en rango de s
    maxErr = -1
    for r in errrange:
        error = abs(zeta(r) - IP(r))
        if error > maxErr:
            maxErr = error
            raux = r

    print(maxErr)
            
    if maxErr < minErr:
        minErr = maxErr
        NminErr = N
        rminErr = raux
    
    N += 1
```
